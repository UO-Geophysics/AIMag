```python
# Make new array with test data columns

train_metadata = np.column_stack((train_phase_ids,train_sta_lats,train_sta_lons,train_source_lats,train_source_lons,train_source_depthskm))
print(train_metadata.shape)

print(train_metadata[0])
print(train_metadata[1])
print(train_metadata[-1])

# np.save('/home/sdybing/mlaapde/traindata_metadata_w_sourcestalocs.txt',)
```

```python
from obspy.geodetics import gps2dist_azimuth, kilometer2degrees

dists_km = []
dists_deg = []

for idx in range(len(train_metadata)):
    
    phase_id = train_metadata[idx,0]
    sta_lat = train_metadata[idx,1]
    sta_lon = train_metadata[idx,2]
    source_lat = train_metadata[idx,3]
    source_lon = train_metadata[idx,4]
    
    distaz = gps2dist_azimuth(source_lat, source_lon, sta_lat, sta_lon)
    dist_m = distaz[0]
    dist_km = dist_m/1000
    dist_deg = kilometer2degrees(dist_km)
    dists_km.append(dist_km)
    dists_deg.append(dist_deg)
    
    if idx == 0:
        print(dists_km)
        print(dists_deg)
    
dists_km_arr = np.array(dists_km)
dists_deg_arr = np.array(dists_deg)

print(dists_km_arr.shape)
print(dists_deg_arr.shape)
```
